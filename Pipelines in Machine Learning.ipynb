```python
from sklearn.pipeline import Pipeline
## feauture scaling
from sklearn.preprocessing import StandardScaler
## model
from sklearn.linear_model import LogisticRegression
```

```python
steps = [('standard_scaler', StandardScaler()), ('classifier', LogisticRegression())]
```

```python
steps
```

```python
Pipeline(steps)
```

```python
pipe = Pipeline(steps)
```

```python
##visualize the pipeline
from sklearn import set_config
set_config(display='diagram')
pipe

```

```python
##creating a dataset
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train


```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

## Complex examples of Columns Transformer

```python
from sklearn.impute import SimpleImputer
```

```python
import numpy as np
```

```python
## numerical processing pipeline
numeric_processor = Pipeline(
    steps=[
        ('imputation_mean', SimpleImputer(missing_values=np.nan, strategy='mean')),
        ('scaler', StandardScaler())
    ]
)

```

```python
numeric_processor
```

```python
## categorical processing pipeline

from sklearn.preprocessing import OneHotEncoder

categorical_processor = Pipeline(
    steps=[
        ('imputation_constant', SimpleImputer(fill_value='missing', strategy='constant')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ]
)

categorical_processor

```

```python
## combine processing techniques
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    [
        ('numeric_processing', numeric_processor, ['age', 'height']),
        ('categorical', categorical_processor, ['gender', 'city'])
    ]
)

preprocessor



```

```python
from sklearn.pipeline import make_pipeline

pipe = make_pipeline(preprocessor, LogisticRegression())

pipe



```
